```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
amount = data["originalMessage"].count()
data.head()
```

```python
print("Total number of messages: ", amount)
```

```python
# show the first 10 rows of the data in a for
originalMessages = data["originalMessage"][:10]
for message in originalMessages:
    print(message)
```

```python
hamming = data["hammingCode"][:10]
for trama in hamming:
    # formated so every 25 bits are separated by a enter
    formated = [trama[i:i+100] for i in range(0, len(trama), 100)]
    print("\n".join(formated))
    print("\n")
```

```python
erroresIndex = data["Error index"][:20]
for trama in erroresIndex:
    if trama ==0:
        print("No se introdujo ningun error a la trama")
    else:
        print("Se introdujo error a la trama en la posicion: ",trama)
```

```python
for i in range(20):
    noise = data["noiseApplied"][i]
    result = data["result of error"][i]
    
    print("noise: ", noise, "\nresult: ", result)
```

```python
# print rows in the data
print("Number of rows in the data = ", len(data))
```

```python
dicc = {
    "true":data["noiseApplied"].value_counts()[True],
    "false":data["noiseApplied"].value_counts()[False]
}

plt.title("Noise applied")
plt.bar(dicc.keys(),dicc.values())
```
